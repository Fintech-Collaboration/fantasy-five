```python
# Save as SQL file
balancer_price_df = pd.read_csv('./data_files/balancer_5_year.csv', 
    parse_dates=True, infer_datetime_format=True, index_col='bal_start_date')
```

```python
Base = declarative_base()
```

```python
# Convert each Dataframe into a Database. Easier for front-end deployment.

# Create the classes
class OHLCV(Base):
    __tablename__ = 'balancer'
    id = Column(Integer, primary_key=True)
    bal_start_date = Column(Date)
    price_open = Column(Float)
    price_high = Column(Float)
    price_low = Column(Float)
    price_close = Column(Float)
    volume_traded = Column(Float)
    trades_count = Column(Integer)
    
engine = create_engine("sqlite:///balancer.sqlite")
conn = engine.connect()

session = Session(bind=engine)
```

```python
Base.metadata.create_all(engine)
```

```python
# Save data as an SQL file.
balancer = balancer_price_df

for idx, row in balancer.iterrows():
    session.add(OHLCV(**row.to_dict()))
    session.commit()
```

```python
# View date row to make sure it came out properly
ohlcv_list = session.query(OHLCV)
for row in ohlcv_list:
    print(row.bal_start_date)
```
