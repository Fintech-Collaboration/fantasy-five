```python
# Save as SQL file
ethereumC_price_df = pd.read_csv('./data_files/ethereumC_5_year.csv', 
    parse_dates=True, infer_datetime_format=True, index_col='etc_start_date')
```

```python
Base = declarative_base()
```

```python
# Convert each Dataframe into a Database. Easier for front-end deployment.

# Create the classes
class OHLCV(Base):
    __tablename__ = 'ethereumC'
    id = Column(Integer, primary_key=True)
    etc_start_date = Column(Date)
    price_open = Column(Float)
    price_high = Column(Float)
    price_low = Column(Float)
    price_close = Column(Float)
    volume_traded = Column(Float)
    trades_count = Column(Integer)
    
engine = create_engine("sqlite:///ethereumC.sqlite")
conn = engine.connect()

session = Session(bind=engine)
```

```python
Base.metadata.create_all(engine)
```

```python
# Save data as an SQL file.
ethereumC = ethereumC_price_df

for idx, row in ethereumC.iterrows():
    session.add(OHLCV(**row.to_dict()))
    session.commit()
```

```python
# View date row to make sure it came out properly
ohlcv_list = session.query(OHLCV)
for row in ohlcv_list:
    print(row.etc_start_date)
```
